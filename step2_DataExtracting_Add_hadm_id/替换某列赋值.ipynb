```python
import pandas as pd
import numpy as np

# 创建lab_df2
data_lab_df2 = {'subject_id': [1, 2, 3, 4,2, 5],
                'hadm_id': [10, np.nan, 30, np.nan,22, 50],
                'chartdate0': ['2022-01-01', '2022-01-02', '2022-01-03', '2022-01-04','2022-01-02', '2022-01-05']}
lab_df2 = pd.DataFrame(data_lab_df2)
lab_df2
```

```python
# 创建after_notnan4
data_after_notnan4 = {'subject_id': [2, 4,3],
                      'hadm_id': [20, 40,30],
                      'chartdate0': ['2022-01-02', '2022-01-04','2022-01-03']}
after_notnan4 = pd.DataFrame(data_after_notnan4)
# 先将after_notnan4的hadm_id列重命名
after_notnan4 = after_notnan4.rename(columns={'hadm_id': 'new_hadm_id'})
after_notnan4
```

```python
# 根据subject_id和chartdate0列合并两个DataFrame
merged_df = pd.merge(lab_df2, after_notnan4[['subject_id', 'chartdate0', 'new_hadm_id']], 
                     on=['subject_id', 'chartdate0'], how='left')
merged_df
```

```python
# 方法一：只给非空的位置重新赋值
# 使用新的hadm_id列替换原来的hadm_id列中的值
merged_df['hadm_id'] = merged_df['hadm_id'].fillna(merged_df['new_hadm_id'])
merged_df
```

```python
# 方法二：可以给选定的位置赋值，但是似乎不是最直观的程序
merged_df['hadm_id'][merged_df['new_hadm_id'].notnull()] = merged_df['new_hadm_id'][merged_df['new_hadm_id'].notnull()]
merged_df
```

```python
# 方法三：最专业的程序
merged_df.loc[merged_df['new_hadm_id'].notnull(), 'hadm_id'] = merged_df['new_hadm_id'][merged_df['new_hadm_id'].notnull()]
merged_df
```
