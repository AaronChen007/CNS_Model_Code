```python
aim1 = pd.read_csv("E:/CNSI_project/demo_and_data_3.0/Processed_data/mimic/TrainingSet/mimic_aim1_dataset0.csv")
aim1
```

```python
# 加载labevent的数据
filename = 'E:/CNSI_project/demo_and_data_3.0/Raw_data/mimic-iv-2.2/hosp/labevents.csv.gz'
with gzip.open(filename, 'rt', encoding='utf-8') as file:
    lab_df = pd.read_csv(file)
lab_df
```

```python
lab_df1 = lab_df[lab_df["subject_id"].isin(aim1["subject_id"])]
lab_df1
```

```python
# 细节：将hadm_id日期推前一天的实验室结果都当成同一个hadm_id，这样可以减少hadm_id的NaN值
lab_df1[lab_df1["subject_id"]==10001217].iloc[55:60]
```

```python
# 从charttime中提取chartdate，并将之转变为秒
lab_df2 = lab_df1.copy()
lab_df2["chartdate"] = lab_df2["charttime"].str[:10]
lab_df2["chartdate0"] = pd.to_datetime(lab_df2["chartdate"], format='%Y/%m/%d')
lab_df2["chartdate0"] = lab_df2["chartdate0"].apply(lambda x: int((x - datetime.datetime(1970,1,1)).total_seconds()))
lab_df2
```

```python
# 若一个case的hadm_id为空，该case的后一天有检测数据且该病人后一天的hadm_id不为空，则将该前一天hadm_id空值赋为后一天的值
befort_isnan = lab_df2[lab_df2['hadm_id'].isnull()]

after_notnan = befort_isnan.copy()
after_notnan["chartdate0"] = after_notnan["chartdate0"]+3600*24
after_notnan
```

```python
# 提取出+1天后有检测数据的情况
# 务必确保after_2cols中的两列不能有重复的数据，要不然根据笛卡尔乘积，将会有很多个重复的结果。
after_2cols = after_notnan[['subject_id','chartdate0']]
after_2cols = after_2cols.drop_duplicates(keep='first')
after_notnan1 = pd.merge(lab_df2, after_2cols, how='inner', on=['subject_id','chartdate0'])
after_notnan1
```

```python
# 提取出after_notnan1的hadm_id非空的情况
after_notnan2 = after_notnan1[after_notnan1['hadm_id'].notnull()]
after_notnan2
```

```python
after_notnan3 = after_notnan2[['subject_id','hadm_id','chartdate0']].drop_duplicates(keep='first')
after_notnan3['chartdate0'] = after_notnan3['chartdate0']-3600*24
after_notnan4 = after_notnan3.rename(columns={'hadm_id': 'new_hadm_id'})
after_notnan4
```

```python
# 根据subject_id和chartdate0列合并两个DataFrame
merged_df = pd.merge(lab_df2, after_notnan4[['subject_id', 'chartdate0', 'new_hadm_id']], 
                     on=['subject_id', 'chartdate0'], how='left')
merged_df
```

```python
# 使用新的hadm_id列替换原来的hadm_id列中的值
merged_df['hadm_id'] = merged_df['hadm_id'].fillna(merged_df['new_hadm_id'])
merged_df
```

```python
# 删除不再需要的列
merged_df = merged_df.drop(columns='new_hadm_id')
# 更新lab_df2
lab_df2 = merged_df
lab_df2
```
